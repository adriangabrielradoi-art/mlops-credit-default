# MLOps Project: Run Everything (Control Panel)

> **This notebook runs the entire Kedro pipeline. No terminal needed.**
> It calls the same `session.run()` that `kedro run` calls under the hood, just
> from inside Jupyter.

## How to use
1. **Select the `.venv-mlops` kernel** (top-right kernel picker). This is the one
   prerequisite. That environment has Kedro and all the libraries.
2. **Run All.** Every pipeline stage executes in order and you see a summary of
   what each produced.

That's it. The per-stage notebooks (`01_...`, `02_...`, `03_...`) are the *detailed
documentation* for the report; this one just runs the whole thing.

## 1. Setup

Bootstrap the Kedro project (find `pyproject.toml`, load settings).

```python
import os
from pathlib import Path
import json
import pandas as pd
from kedro.framework.session import KedroSession
from kedro.framework.startup import bootstrap_project

# Project root works whether this notebook is run from notebooks/ or the root.
PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()

# IMPORTANT: run from the project root so relative paths (e.g. the MLflow
# sqlite:///mlflow.db tracking store) resolve to the SAME place the terminal
# `kedro run` uses — not a stray copy under notebooks/.
os.chdir(PROJECT_ROOT)

DATA = PROJECT_ROOT / "data"
bootstrap_project(PROJECT_ROOT)
print("Working dir:", Path.cwd().name, "| Python:", "venv-mlops" if "venv-mlops" in os.sys.executable else os.sys.executable)
print("Project bootstrapped:", PROJECT_ROOT.name)
```

## 2. Run the full pipeline

`session.run()` runs the default pipeline. Every stage, in dependency order
(data_quality to data_cleaning to data_feat_engineering to ... as they are built).
This is **identical** to running `kedro run` in a terminal.

```python
with KedroSession.create(project_path=PROJECT_ROOT) as session:
    session.run()   # __default__ = the full sequence
print("\nFull pipeline run complete.")
```

## 3. What each stage produced

A quick summary read from the artifacts the run just wrote.

```python
# data_quality
ge = json.load(open(DATA / "08_reporting" / "ge_report.json"))
print(f"data_quality      : {ge['n_expectations']} expectations, {ge['n_failed']} failed -> passed={ge['success']}")

# data_cleaning
cleaned = pd.read_parquet(DATA / "03_primary" / "cleaned_data.parquet")
print(f"data_cleaning     : cleaned table {cleaned.shape}")

# data_feat_engineering (+ Hopsworks feature store)
ft = pd.read_parquet(DATA / "04_feature" / "feature_table.parquet")
print(f"data_feat_eng     : feature table {ft.shape}")
fs = json.load(open(DATA / "08_reporting" / "feature_store_status.json"))
print(f"feature_store     : connected={fs['connected']}, groups={len(fs['feature_groups_written'])}, rows={fs['rows_written']}")

# data_split
tr = pd.read_parquet(DATA / "05_model_input" / "X_train.parquet").shape
va = pd.read_parquet(DATA / "05_model_input" / "X_val.parquet").shape
te = pd.read_parquet(DATA / "05_model_input" / "X_test.parquet").shape
print(f"data_split        : train={tr}, val={va}, test={te}")

# model_train
tm = json.load(open(DATA / "08_reporting" / "training_metrics.json"))
best = max(tm, key=lambda k: tm[k]["roc_auc"])
print(f"model_train       : {len(tm)} models, best={best} (roc_auc={tm[best]['roc_auc']:.4f})")

# model_selection (champion + MLflow registry)
ci = json.load(open(DATA / "08_reporting" / "champion_info.json"))
print(f"model_selection   : champion={ci['champion_family']} ({ci['primary_metric']}={ci['value']:.4f}), registry v{ci.get('registered_version','n/a')}")

# model_predict
dl = json.load(open(DATA / "08_reporting" / "decision_log.json"))
print(f"model_predict     : scored {dl['n_scored']} rows, predicted default rate={dl['predicted_default_rate']:.3f}")

# data_drifts
dr = json.load(open(DATA / "08_reporting" / "drift_summary.json"))
print(f"data_drifts       : control batch {dr['clean']['n_features_drifted']} drifted | biased batch {dr['drifted']['n_features_drifted']} drifted")
```
